```python

a = int(input("Enter your first numbe "))
b = int(input("Enter your second number"))
def swap(a, b):

    print("select the method")
    print("1. Using temp variables")
    print("2. Using add and sub")
    print("3. Using tuple swapping")

    c = int(input("enter your choice "))

    if(c == 1):

        print("method 1 selected")
        temp = a
        a = b
        b = temp
        print(f"a={a},b={b}")

    elif(c == 2):
        print("method 2 selected")
        a = a + b
        b = a - b
        a = a - b
        print(f"a={a},b={b}")

    elif(c == 3):
        print("method 3 selected")
        a, b = b, a
        print(f"a={a},b={b}")

    else:
        print("Invalid choice")


swap(a, b)
```
